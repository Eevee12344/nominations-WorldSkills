```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

def parse_habr_page(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Найдем блок с данными о номинациях и компаниях
    start_marker = "Лауреаты «Премии Рунета 2023»"
    end_marker = "В Народном голосовании определён лауреат в номинации «Лучший социальный проект Рунета»: Добровольческий поисково-спасательный отряд «ЛизаАлерт». По результатам зрительского голосования второе место в Народном голосовании пользователи отдали Всероссийскому проекту «Разговоры о важном». В голосовании приняли участие более 100 000 пользователей."
    start_block = soup.find(text=re.compile(start_marker))
    end_block = soup.find(text=re.compile(end_marker))

    if start_block and end_block:
        nominations_data = []
        current_block = start_block.find_next('p')
        while current_block and current_block != end_block:
            if current_block.name == 'p':
                nomination_title = current_block.text.strip()
                companies = re.findall(r'(?<=▪ )[\w\s\.\(\)«»-]+', current_block.text)
                for company in companies:
                    nominations_data.append({'Номинация': nomination_title, 'Компания': company})
            current_block = current_block.find_next('p')
        
        return nominations_data
    else:
        print("Start or end markers not found.")
        return []

habr_url = 'https://habr.com/ru/news/781170/'
nominations_data = parse_habr_page(habr_url)

# Проверим данные
df = pd.DataFrame(nominations_data)
print(df.head())

# Сохраняем данные в CSV файл
df.to_csv('habr_nominations.csv', index=False)
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re

# Загрузка данных
df = pd.read_csv('habr_nominations.csv')  # Замените 'your_data.csv' на путь к вашему файлу данных

# Обработка пропусков данных
df.dropna(inplace=True)

# Очистка текстов от лишних символов
def clean_text(text):
    text = re.sub(r'[^\w\s]', '', text)  # Удаляем все символы, кроме букв и пробелов
    return text

df['cleaned_text'] = df['Номинация'].apply(clean_text)

# Токенизация текстов
df['tokenized_text'] = df['cleaned_text'].apply(word_tokenize)

# Лемматизация текстов
lemmatizer = WordNetLemmatizer()
def lemmatize_text(tokens):
    return [lemmatizer.lemmatize(token) for token in tokens]

df['lemmatized_text'] = df['tokenized_text'].apply(lemmatize_text)

# Удаление спецсимволов из текстов
def remove_special_characters(tokens):
    return [re.sub(r'[^a-zA-Z0-9]', '', token) for token in tokens]

df['cleaned_text'] = df['lemmatized_text'].apply(remove_special_characters)

# Выделение значимых частей речи
def extract_pos(tokens):
    return [word for word, pos in nltk.pos_tag(tokens) if pos.startswith('NN') or pos.startswith('VB')]

df['pos_tags'] = df['lemmatized_text'].apply(extract_pos)

# Удаление стоп-слов
stop_words = set(stopwords.words('russian'))  # Замените 'english' на язык, на котором написан ваш текст

def remove_stopwords(tokens):
    return [token for token in tokens if token not in stop_words]

df['cleaned_text'] = df['cleaned_text'].apply(remove_stopwords)

# Сохраняем обработанные данные
df.to_csv('processed_data.csv', index=False)
```
